```python
import numpy as np
import matplotlib.pyplot as plt
# Optional: SymPy for symbolic mathematics (if you want to demonstrate vector equations)
import sympy as sp
from numpy.ma.core import angle
```

# Vector Operations

```python
a = np.array([2, 8])
b = np.array([6, 4])
```

## Vector Addition

```python
c = a + b

plt.quiver(0, 0, a[0], a[1], angles='xy', scale_units='xy', scale=1, color='g', label='a')
plt.quiver(0, 0, b[0], b[1], angles='xy', scale_units='xy', scale=1, color='g', label='b')
plt.quiver(0, 0, c[0], c[1], angles='xy', scale_units='xy', scale=1, color='r', label='c')

plt.xlim(-1,10)
plt.ylim(-1,14)

plt.grid()
plt.legend()
plt.show()
```

## Scalar Multiplication

```python
scalar = 3
scaled_vector = scalar * b

plt.quiver(0,0, a[0], a[1], angles='xy', scale_units='xy', scale=1, color='g', label='Original')
plt.quiver(0, 0, scaled_vector[0], scaled_vector[1], angles='xy', scale_units='xy', scale=1, color='r', label='Scaled')

plt.xlim(-1,scaled_vector[0] + 2)
plt.ylim(-1,scaled_vector[1] + 2)

plt.text(a[0], a[1], f'a = {a}',  ha='left', va='top')
plt.text(scaled_vector[0], scaled_vector[1], f'a = {scaled_vector}',  ha='right', va='top')

plt.legend()
plt.grid()
plt.show()
```

## Dot or Scalar Product

### Mathematical Definition:

$$ \mathbf{a} = [a_1, a_2, \ldots, a_n] $$ 
$$ \mathbf{b} = [b_1, b_2, \ldots, b_n] $$

$$ \mathbf{a} \cdot \mathbf{b} = a_1b_1 + a_2b_2 + \ldots + a_nb_n $$

### Geometric Interpretation:

$$ \mathbf{a} \cdot \mathbf{b} = ||\mathbf{a}|| \cdot ||\mathbf{b}|| \cdot \cos(\theta) $$


### Properties:
1. **Commutative**: $$ \mathbf{a} \cdot \mathbf{b} = \mathbf{b} \cdot \mathbf{a} $$
2. **Distributive**: $$ \mathbf{a} \cdot (\mathbf{b} + \mathbf{c}) = \mathbf{a} \cdot \mathbf{b} + \mathbf{a} \cdot \mathbf{c} $$
3. **Scalar**: The result is a scalar, not a vector.

### Example:
$$ \mathbf{a} = [2, 3] $$
$$ \mathbf{b} = [5, 4] $$

$$ \mathbf{a} \cdot \mathbf{b} = (2 \times 5) + (3 \times 4) = 10 + 12 = 22 $$

```python
dot_product = np.dot(a, b)

plt.quiver(0, 0, a[0], a[1], angles='xy', scale_units='xy', scale=1, color='g', label='a')
plt.quiver(0, 0, b[0], b[1], angles='xy', scale_units='xy', scale=1, color='r', label='b')

# Annotate the dot product value
plt.text(0, -0.5, f'Dot Product = {dot_product}', fontsize=12, color='black')

plt.xlim(-1,b[0] + 2)
plt.ylim(-1,b[1] + 2)

plt.title('Dot Product of Vectors a and b')
plt.grid()
plt.legend()
plt.show()
```
